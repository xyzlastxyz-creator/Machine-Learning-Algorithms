```python
from sklearn.datasets import load_iris  # Load the built-in Iris dataset
from sklearn.tree import DecisionTreeClassifier, plot_tree  # For creating and visualizing decision trees
from sklearn.metrics import confusion_matrix, accuracy_score  # For checking model performance
import matplotlib.pyplot as plt  # For plotting graphs
import seaborn as sns  # For making stylish plots like heatmaps
```

```python
# Load the Iris dataset : Setosa, Versicolor, Virginica
# Each flower is described by 4 features: Sepal Length, Sepal Width, Petal Length, Petal Width
iris = load_iris()

# Separate the data into features (X) and labels (y)
X = iris.data 
y = iris.target  # Output labels (flower species) â€“ 0, 1, 2 setosa, versinica, versicolor
```

```python
# Create a Decision Tree classifier
# criterion='entropy' tells the tree to use entropy to decide splits
# max_depth=3 limits the tree depth to avoid overfitting
# random_state=42 ensures the same result every time (for reproducibility)
clf = DecisionTreeClassifier(criterion='entropy', max_depth=7, random_state=42)
```

```python
# Train the Decision Tree model using the input data
clf.fit(X, y)

# Predict the classes of the same training data (just to demo performance)
y_pred = clf.predict(X)
print (y_pred)
```

```python
# Check how accurate the predictions are compared to actual labels
accuracy = accuracy_score(y, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
# Create the confusion matrix to compare actual vs predicted values
cm = confusion_matrix(y, y_pred)
print ("Confusion matrix : ")
print (cm)
```

```python
# Plot the confusion matrix as a heatmap
plt.figure(figsize=(6, 4))  # Set size of the plot
sns.heatmap(cm, annot=True, cmap="Blues",  # Show numbers inside boxes, use blue shades
            xticklabels=iris.target_names,  # Label x-axis with flower names
            yticklabels=iris.target_names)  # Label y-axis with flower names
plt.xlabel("Predicted")  # Label for x-axis
plt.ylabel("Actual")     # Label for y-axis
plt.title("Confusion Matrix")  # Title of the plot
#plt.savefig("iris_confusion_fig2.jpg")  # Save the plot as an image
plt.show()  # Show the plot
```

```python
# Visualize the actual structure of the trained Decision Tree
plt.figure(figsize=(15, 9))  # Create a big figure so the tree fits well
plot_tree(clf,
          feature_names=iris.feature_names,  # Show feature names on the tree
          class_names=iris.target_names,     # Show flower class names
          filled=True,                       # Fill color in the boxes for better understanding
          rounded=True)                      # Make the boxes rounded
plt.title("Decision Tree Visualization")     # Title of the tree diagram
#plt.savefig("iris_confusion_fig.jpg")        # Save the tree diagram as an image
plt.show()                                   # Display the tree
```
